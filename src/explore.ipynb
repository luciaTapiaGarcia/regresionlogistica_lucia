### Análisis y tratamiento de outliers

He elegido estas cuatro porque son las que realmente tienen outliers "problemáticos". Las variables macroeconómicas (emp.var.rate, euribor3m, cons.price.idx, etc.) las he dejado fuera porque sus valores raros no son errores, simplemente reflejan distintos momentos económicos. Con pdays pasa algo parecido: ese 999 que aparece tanto no es un outlier, es la forma que tiene el dataset de decir "este cliente no fue contactado antes". En cambio, en duration, campaign, age y previous sí vemos valores que se salen demasiado de lo normal y que podrían liar al modelo, así que tiene sentido recortarlos.

#### Duration

```python
duration_stats = total_data["duration"].describe()
duration_stats
```

```python
iqr = duration_stats["75%"] - duration_stats["25%"]
upper_limit = duration_stats["75%"] + 1.5 * iqr
lower_limit = duration_stats["25%"] - 1.5 * iqr

print(f"Límite superior: {round(upper_limit, 2)}, Límite inferior: {round(lower_limit, 2)}, IQR: {round(iqr, 2)}")

outliers = total_data[total_data["duration"] > upper_limit]
print(f"Outliers en duration: {len(outliers)} ({round(len(outliers)/len(total_data)*100, 2)}%)")

total_data[total_data["duration"] > upper_limit]

print(f"Valores nulos en duration: {total_data['duration'].isnull().sum()}")
```

La variable duration tiene una media de 258 segundos pero una mediana de 180, lo que confirma el sesgo a la derecha que ya observamos en el histograma. El límite superior IQR se sitúa en 644.5 segundos, y hay 2.963 registros por encima de ese valor, representando un 7.2% del total. Son llamadas extremadamente largas que podrían distorsionar el modelo. No hay valores nulos.

### Transformación de variables categóricas

```python
import numpy as np
import pandas as pd

# Codificamos las variables categóricas a numéricas
total_data["job_n"] = pd.factorize(total_data["job"])[0]
total_data["marital_n"] = pd.factorize(total_data["marital"])[0]
total_data["education_n"] = pd.factorize(total_data["education"])[0]
total_data["default_n"] = pd.factorize(total_data["default"])[0]
total_data["housing_n"] = pd.factorize(total_data["housing"])[0]
total_data["loan_n"] = pd.factorize(total_data["loan"])[0]
total_data["contact_n"] = pd.factorize(total_data["contact"])[0]
total_data["month_n"] = pd.factorize(total_data["month"])[0]
total_data["day_of_week_n"] = pd.factorize(total_data["day_of_week"])[0]
total_data["poutcome_n"] = pd.factorize(total_data["poutcome"])[0]

# Lista de todas las variables que usaremos para entrenar el modelo
num_variables = [
    "age",
    "job_n",
    "marital_n",
    "education_n",
    "default_n",
    "housing_n",
    "loan_n",
    "contact_n",
    "month_n",
    "day_of_week_n",
    "duration",
    "campaign",
    "pdays",
    "previous",
    "poutcome_n",
    "emp.var.rate",
    "cons.price.idx",
    "cons.conf.idx",
    "euribor3m",
    "nr.employed"
]

# X con todas las variables predictoras
X = total_data[num_variables].apply(pd.to_numeric, errors="coerce")

# Variable objetivo
y = total_data["y"]
```

PAsamos las categoricas a numericas y unimos todas para darle el valor a la X e Y nuestra variable objetivo

### División en train y test

```python
from sklearn.model_selection import train_test_split

# Dividimos los datos en entrenamiento (80%) y test (20%)
# random_state se fija para que la división sea reproducible (si alguien más ejecuta el código, obtendrá la misma división)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```

### Crear pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif

# Creamos un pipeline para imputar los valores faltantes con la mediana 
# y luego seleccionar las 5 mejores características según su relación con la variable objetivo
pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("selector", SelectKBest(score_func=f_classif, k=5))  # f_classif para clasificación
])
```

### Entrenar selección de características

```python
# Entrenamos el pipeline solo con los datos de entrenamiento 
# por que queremos evitar  que el modelo tenga información de los datos de test (lo que se llama "data leakage") y así poder evaluar su rendimiento de manera más realista en datos nuevos
X_train_sel = pipe.fit_transform(X_train, y_train)
```

### Optimizacion

#### Encontrar los mejores hiperparámetros

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Modelo base
model = LogisticRegression(
    random_state=42,
    max_iter=1000
)

# Hiperparámetros
hyperparams = {
    "C": np.logspace(-3, 2, 20),
    "penalty": ["l1", "l2"],
    "solver": ["liblinear"],
    "class_weight": [None, "balanced"]
}

# Random Search
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=hyperparams,
    n_iter=15,              # SOLO prueba 15 combinaciones
    scoring="recall",
    cv=3,                   # más rápido que 5
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Entrenamiento
random_search.fit(X_train_sel, y_train)

# Mejores parámetros
print("Mejores hiperparámetros:")
print(random_search.best_params_)
```

```python
# Creamos un nuevo modelo con los mejores hiperparámetros encontrados
best_model = random_search.best_estimator_
# Hacemos predicciones con el modelo optimizado
y_pred = best_model.predict(X_test_sel)
```

```python
#Comprobamos la precisión del modelo optimizado
accuracy_score(y_test, y_pred)
```

```python
# Y lo dibujamos con una matriz de confusión para ver cómo ha mejorado el modelo optimizado respecto al modelo base
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm)

plt.figure(figsize=(4, 4))
sns.heatmap(cm_df, annot=True, fmt="d", cbar=False, cmap="RdPu")   # colores rosas
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.tight_layout()
plt.show()
```
